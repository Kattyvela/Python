# Análisis de datos de participantes en las Juegos Olímpicos
En este informe se detallan ciertos análisis sobre los datos de los participantes en Juegos Olímpicos. Partiendo de un análisis global a un análisis de las participaciones en juegos olímpicos de Ecuador.

## Descripción del conjunto de datos
La información fue tomada de la página [Kaggle](https://www.kaggle.com/heesoo37/120-years-of-olympic-history-athletes-and-results) y se trata de un conjunto de datos históricos de los juegos olímpicos modernos. La información incluye todas las participaciones desde Atenas 1896 hasta Río 2016 de juegos olímpicos de veranos e invierno.

La información que contiene el archivo es:

-  **ID**: Identificador único del deportista
-  **Name**: Nombre del deportista
-  **Sex**: Género del deportista
-  **Age**: Edad del deportista al momento de su participación
-  **Height**: Estatura del deportista al momento de su participación
-  **Weight**: Peso del deportista al momento de su participación
-  **Team**: País al que representó el deportista
-  **NOC**: Abreviación del país (tres letas)
-  **Games**: Concatenación del año y estación del evento
-  **Year**: Año realización del evento
-  **Season**: Estación del evento
-  **City**: Ciudad donde se realizó el evento
-  **Sport**: Deporte de participación del deportista
-  **Event**: Prueba 
-  **Medal**: Medalla obtenida por el deportista

## Pariticipantes en Juegos Olímpicos Históricamente
El conjunto de datos contiene la información de los juegos olímpicos de verano e invierno, la información de participantes y deportes se distribuye de la siguiente forma:

![Participantesxest](.\Img\EstPart.PNG)

Como se observa en el gráfico, del total de registros el 86,03% corresponden a juegos olímpicos de verano y el 13,97% a juegos olímpicos de invierno, esto se debe a que en los juegos de verano existen más deportes y disciplinas en las que un deportista puede participar, en juegos de verano los deportistas pueden participar en 52 deportes diferentes mientras que en juego de invierno solo existen 17 deportes.

Para este análisis se tomará en cuenta solo la información de juegos olímpicos de verano debido a que en estos existen más participantes y más deportes. Además para el análisis de los datos correspondientes a Ecuador solo tenemos participaciones en juegos olímpicos de verano.

Con el conjunto de datos reducido a juegos olímpicos de verano haremos un análisis de número de participantes por año y de ellos veremos cuantos son hombres y cuantas mujeres.

![Participantesxañoygen](.\Img\PartxAñoyGen.PNG)

Del gráfico anterior podemos deducir que con el paso del tiempo se ha ido incrementando el número de participantes por edición de juegos olímpicos. La participación de hombres es mayor que la de mujeres en todas las ediciones, a pesar de que las mujeres han aumentado su porcentaje de participación en los últimos años no llega a ser igual que la de los hombres.

Además, se puede observar que a partir de los años 90 existe un mayor número de deportes por edición.

## Ecuador en Juegos Olímpicos
Para esta sección se redujo el conjunto de datos a únicamente los deportistas que han representado al Ecuador en cualquier edició de Juegos Olímpicos. 

![Ecu](.\Img\Ecu.PNG)

Ecuador participó por primera vez en Juegos Olímpicos en el año 1924 y luego en 1968 desde donde ha tenido representantes en todas las ediciones de Juegos Olímpicos. 
Generalmente el número de participantes de género masculino supera a las de género femenino, con una edición en particular, Barcelona 1992, donde existen más mujeres que hombres participantes.

![PartiEdiGen](.\Img\PartEdiGen.PNG)

Tal como sucede con el conjunto de datos total, con el paso del tiempo Ecuador ha ido aumentando el número de participantes y de deportes en los que tiene representantes, siendo el 2016 la edición en la que más representantes ha tenido con un total de 37 participantes, 20 hombres y 15 mujeres en 12 deportes.

![EcuMed](.\Img\DepEdi.PNG)


El deporte con mayor porcentaje de participación de Ecuador en los juegos olimpicos es en Atletismo con un 31,44%, seguido de la natación y el ciclismo con 17,42% y 9,85% respectivamente.


![EcuDep](.\Img\EcuDep.PNG)


También se realizó un análisis por edad de todos los deportistas que han participado en juegos olímpicos, del cual se pudo concluir que la mediana de la edad en mujeres es mayor que en hombres participantes.


![EcuMed](.\Img\MedEdad.PNG)


En cuanto al medallero, Ecuador registra únicamente 2 medallas, una de oro en Atlanta 1996 obtenida por Jefferson Pérez en la prueba 20Km caminata y la otra en Beijing 2008 también obtenida por Jefferson Pérez en la misma prueba.

![EcuMed1](.\Img\Med.PNG)

Por último se realizó un análisis de número de participantes de la última edición de Juegos Olímpicos, Río 2016, agrupado por deporte y género, del cual se concluye que para la mayoría de deportes el porcentaje de participación del género masculino es mayor que la del género femenino, sin embrago, existen deportes como el levantamiento de pesas para el que existen más mujeres que hombres participantes e incluso en triatlon solo existió 1 representante de género femenino.

![DepGen](.\Img\DepGen2016.PNG)

Con el análisis realizado se concluye que tanto a nivel global como en Ecuador conforme ha pasado el tiempo ha crecido el número de participantes en juegos olímpicos, esto puede acreditarse al aumento de competencias consideradas como clasificatorias a juegos olímpicos y mayormente al uso de la tecnología como parte de la planificación de entrenamientos en cada deporte y disciplina. 
También se puede decir que el número de medallas obtenidas guarda cierta relación con el porcentaje de desarrollo tecnológico del país, siendo así varios de los países más desarrolldos los que aparecen primeros en el medallero general.
